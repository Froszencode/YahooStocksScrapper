```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
custom_useragent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'
options = Options()
options.add_argument(f"--user-agent={custom_useragent}")
driver = webdriver.Chrome(options=options)
driver.maximize_window()

#explicit wait
wait=WebDriverWait(driver,5)


# fucnction checks if the webpage is loaded properly.
def wait_for_page_to_load(driver,wait): 
    try:
        wait.until(
            lambda d : d.execute_script("return document.readyState") == "complete"
        )
        time.sleep(3)
        page_title = driver.title
        print(f"The \"{page_title}\" is fully loaded.")
    except:
        time.sleep(3)
        page_title = driver.title
        print(f"The \"{page_title}\" page is not loaded within the given time.")


url="https://finance.yahoo.com/"
driver.get(url)
wait_for_page_to_load(driver,wait)



#hovering over markets menu
action=ActionChains(driver)
markets_menu = wait.until(
    EC.presence_of_element_located((By.XPATH,'//*[@id="navigation-container"]/ol/li[3]/a/div'))
)
action.move_to_element(markets_menu).perform()

#hovering over stocks menu
stocks_menu = wait.until(
    EC.presence_of_element_located((By.XPATH,'/html[1]/body[1]/div[2]/header[1]/div[1]/nav[1]/ol[1]/li[3]/ol[1]/li[1]/a[1]'))
)
action.move_to_element(stocks_menu).perform()

#clicking on 'trending' button
trending_btn = wait.until(
    EC.element_to_be_clickable((By.XPATH,'/html[1]/body[1]/div[2]/header[1]/div[1]/nav[1]/ol[1]/li[3]/ol[1]/li[1]/ol[1]/li[4]/a[1]/span[1]'))
)
action.click(trending_btn).perform()
wait_for_page_to_load(driver,wait)

#opeing most active page
most_active_btn = wait.until(
    EC.element_to_be_clickable((By.XPATH,'/html[1]/body[1]/div[2]/main[1]/section[1]/section[1]/section[1]/section[1]/section[1]/div[1]/div[1]/div[1]/a[1]'))
)
action.click(most_active_btn).perform()
wait_for_page_to_load(driver,wait)

#scrapping stocks data
data=[]
while True:
    #scraping
    wait.until(
        EC.presence_of_element_located((By.TAG_NAME,"table"))
    )
    rows = driver.find_elements(By.CSS_SELECTOR,"table tbody tr")
    for row in rows:
        values = driver.find_elements(By.TAG_NAME,"td")
        stock={
            "name":values[1].text,
            "symbol":values[0].text,
            "price":values[3].text,
            "change":values[4].text,
            "volume":values[6].text,
            "market_cap":values[8].text,
            "pe_ratio":values[9].text,
        }
        data.append(stock)
    #click next
    try:
        next_btn = wait.until(
            EC.element_to_be_clickable((By.XPATH,'//*[@id="main-content-wrapper"]/section[1]/div/div[4]/div[3]/button[3]'))
        )
    except:
        print("next btn is not clicable, we have navigated through all the pages")
        break
    else:
        next_btn.click()
        time.sleep(1)
driver.quit()
```
